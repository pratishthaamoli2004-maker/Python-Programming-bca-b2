### 1) Find a factorial of given number.

```python
num = int(input("Enter a number: "))
fact = 1

for i in range(1, num + 1):
    fact = fact * i
print("Factorial =, fact")
```

### 2) Find whether the given number is armstrong number.

```python
num = int(input("Enter a  number: "))
temp = num
sum = 0
digits = len(str(num))

while temp > 0:
    digit = temp % 10
    sum += digit ** digits
    temp //= 10
    
    if sum== num:
        print("armstrong number")
    else:
        print("not armstrong number")
```

### 3) Print Fibonacci series up to given term .

```python
n = int(input("Enter number of terms: "))
a = 0
b = 1
for i in range(n):
    print(a, end=" ")
    c = a + b
    a = c
    b = c
```

### 4) Write a program to find if given number is prime number or not.

```python
num = int(input("Enter a number: "))
count = 0

for i in range(1, num + 1):
    if num % i == 0:
        count += 1
        
        if count == 2:
            print("Prime Number")
        else:
            print("Not Prime Number")
```

### 5) Check whether given number is palindrome or not.

```python
num = input("Enter a number: ")
original = num

reverse = ""

for i in original:
    reverse = i + reverse

if original == reverse:
    print("The given number is a Palindrome.")
else:
    print("The given number is not a Palindrome.")
```

### 6) Write a program to print sum of digits.

```python
num = int(input("Enter a number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum += digit
    num //= 10

print("Sum of digits =", sum)
```

### 7)  Count and print all numbers divisible by 5 or 7 between 1 to 100.

```python
count = 0

for i in range(1, 101):
    if i % 5== 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1
        
        print("\nTotal count =", count)
```

### 8) Convert all lower cases to upper case in a string


```python
text = input("Enter a string: ")
print("\nOriginal String:", text)

uppercase_text = text.upper()
print("Converted String in Uppercase: ", uppercase_text)

if text == uppercase_text:
    print("The string was already in uppercase: ")
else:
    print("Lowercase Letters have been converted to uppercase successfully: ")

```

### 9) Print all prime numbers between 1 and 100 

```python
for num in range(1, 101):
    count = 0

    for i in range(1, num + 1):
        if num % i == 0:
            count += 1

    if count == 2:
        print(num, end=" ")
```

### 10) Print the table for a given number:
<br>5 * 1 =  5</br>
<br>5 * 2 = 10</br>

```python
num = int(input("Enter a number to generate its multiplication table: "))
print("\n------------------------------")
print("Multiplication Table of", num)
print("\n------------------------------")

for i in range(1, 11):
    result = num * i
    print(num, "x", i, "=", result)
print("\n------------------------------")
print("End of table")
```
