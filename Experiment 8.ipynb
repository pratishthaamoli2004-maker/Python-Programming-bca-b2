### **1) Add few names, one name in each row, in "name.txt file."**

**a. Count no of names**

**b. Count all names starting with vowel**

**c. Find longest name**


### **a.**

```python
f = open("notes.txt", "r")
count = 0
for line in f:
    if line.strip() != "":
        count += 1
f.close()
print("Total names:", count)
```

### **b.**

```python
count = 0
with open("notes.txt", "r") as f:
    for line in f:
        if line and line[0].lower() in 'aeiou':
            count += 1
print("Names starting with vowel:", count)
```

### **c.**

```python
f = open("notes.txt", "r")
longest = ""
for line in f:
    name = line.strip()
    if len(name) > len(longest):
        longest = name
f.close()
print("Longest name:", longest)
```

### **2) Store integers in a file.**

**a. Find the max number**

**b. Find average of all numbers**

**c. Count number of numbers greater than 100**


### **a.**

```python
f = open("num.txt", "r")
nums = [int(x.strip()) for x in f if x.strip() != ""]
f.close()

if len(nums)== 0:
    print("File is empty")
else:
    maxi = nums[0]
    for n in nums:
        if n > maxi:
            maxi = n
    print("Maximum:", maxi)
```

### **b.**

```python
f = open("num.txt", "r")
nums = [int(x) for x in f]
f.close()
total = sum(nums)
avg = total / len(nums)
print("Average:", avg)
```

### **c.**

```python
f = open("num.txt", "r")
count = 0
for line in f:
    if int(line) > 100:
        count += 1
f.close()
print("Numbers > 100:", count)
```

### **3) Assume a file city.txt with details of 5 cities in given format (city names population(in lakhs) area(in sq KM) ):**

**Example:**

**Dehradun 5.78 308.20**

**Delhi 190 1484**

**……………**

**Open file city.txt and read to:**

**a. Display details of all cities**

**b. Display city names with population more than 10Lakhs**

**c. Display sum of areas of all cities**

### **a.**

```python
f = open("city.txt", "r")
for line in f:
    print(line.strip())
f.close()
```

### **b.**

```python
f = open("city.txt", "r")
for line in f:
    data = line.split()
    if float(data[1]) > 10:
        print(data[0])
f.close()
```

### **c.**


```python
f = open("city.txt", "r")
total = 0
for line in f:
  data = line.split()
  total += float(data[2])
f.close()
print("Total Area:", total)
```

### **4) Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform  integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.**

**Sample input**

**1 0**

**2 $**

**3 1**

**Sample Output:**

**Error Code: integer division or modulo by zero**

**Error Code: invalid literal for int() with base 10: '$' 3**

```python
n = int(input())
for i in range(n):
  try:
      a, b = input().split()
      print(int(a)//int(b))
  except Exception as e:
     print("Error Code:", e)
```

### **5) Create multiple suitable exceptions for a file handling program.**

```python
try:
  with open("data.txt", "r") as myf:
    count = 0
    total= 0
    for line in myf:
        parts = line.strip().split()
        if len(parts) < 2:
            continue
        num = int(parts[1])
        count += 1
        total += num
    print("Average:", total / count)
except FileNotFoundError:
    print("Error: File not found.")
except ValueError:
    print("Error: File contains non-numeric data.")
except ZeroDivisionError:
    print("Error: File is empty.")
```
