```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        if cat not in d:
            d[cat] = {}
            if word not in d[cat]:
                d[cat][word] = 0
                d[cat][word] +=1
                return d
```

```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
        if x not in seen:
            seen.add(x)
            return list(seen)
```

### question 5


```python
def moving_average(stream, w):
    result = []
    window_sum = 0
    
    for i in range(len(stream)):
        window_sum += stream[i]
        if i >= w:
            window_sum -= stream[i - w]
            current_window = min(i + 1, w)
            avg = window_sum / current_window
            result.append(avg)
```
