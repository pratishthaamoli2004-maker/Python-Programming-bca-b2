### 1. Check whether a given number is divisible by 3 and 5 both.

```python
num = int(input("Enter number: "))

if num % 3 == 0 and num % 5 == 0:
    print("Divisible by both 3 and 5")
else:
    print("Not divisible by both 3 and 5")
```

### 2. Check whether a given number is multiple of five or not.

```python
num = int(input("Enter number: "))

if num % 5 == 0:
    print("Multiple of 5")
else:
    print("Not a multiple of 5")
```

### 3. Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Greatest is", a)
elif b > a:
    print("Greatest is", b)
else:
    print("Numbers are equal")
```

### 4. Find the greatest among three numbers assuming no two values are same.

### 5. Check whether the quadratic equation has real roots or imaginary roots. Display the roots.

```python
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

D = b*b - 4*a*c

if D > 0:
    print("Real and different roots")
    r1 = (-b + math.sqrt(D))/(2*a)
    r2 = (-b - math.sqrt(D))/(2*a)
    print("Roots:", r1, r2)

elif D == 0:
    print("Real and equal roots")
    r = -b/(2*a)
    print("Root:", r)

else:
    print("Imaginary roots")
```

### 6. Find whether a given year is a leap year or not.

```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap year")
else:
    print("Not leap year")
```

### 8. Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and calculate the percentage.
<br>CGPA=percentage/10</br>
<br>CGPA range:</br>
<br>0 to 3.4 -> F</br>
<br>3.5 to 5.0->C+</br>
<br>5.1 to 6->B</br>
<br>6.1 to 7-> B+</br>
<br>7.1 to 8-> A</br>
<br>8.1 to 9->A+<br>
<br>9.1 to 10-> O (Outstanding)</br>
<p>Sample Gradesheet</p>
<br>Name: Rohit Sharma</br>
<br>Roll Number: R17234512 SAPID: 50005673</br>
<br>Sem: 1 Course: B.Tech. CSE AI&ML</br>
<br>Subject name: Marks</br>
<br>PDS: 70</br>
<br>Python: 80</br>
<br>Chemistry: 90</br>
<br>English: 60</br>
<br>Physics: 50</br>
<br>Percentage: 70%</br>
<br>CGPA:7.0</br>
<br>Grade: A</br>

```python
name = "Rohit Sharma"
roll = "R17234512"
sapid = "50005673"
sem = 1
course = "B.Tech CSE AI&ML"

pds = 70
python = 80
chem = 90
eng = 60
phy = 50

total = pds + python + chem + eng + phy
percentage = total / 5

cgpa = percentage / 10

if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O (Outstanding)"

print("Name:", name)
print("Roll Number:", roll)
print("SAPID:", sapid)
print("Semester:", sem)
print("Course:", course)

print("\nMarks:")
print("PDS:", pds)
print("Python:", python)
print("Chemistry:", chem)
print("English:", eng)
print("Physics:", phy)

print("\nPercentage:", percentage, "%")
print("CGPA:", cgpa)
print("Grade:", grade)
```
