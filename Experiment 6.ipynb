### **1) Scan n values in range 0-3 and print the number of times each value has occurred.**

```python
n=int(input("Enter n: "))
a=[int(input()) for i in range(n)]
for i in range(4):
    print(i,"occurred:",a.count(i),"times")
```

### **2) Create a tuple to store n numeric values and find the average of all values.**

```python
n=int(input("Enter n: "))
t=tuple(float(input()) for i in range(n))
print("Tuple:",t)
print("Average:",sum(t)/n)
```

### **3) WAP to input a list of scores for N students in a list data type. Find the score ofthe runner-up and print the output.**

**Sample Input**

**N = 5**

**Scores= 2 3 6 6 5**

**Sample output**

**5**

**Note: Given list is [2, 3, 6, 6, 5]. The maximum score is 6, second maximum is 5. Hence, we print 5 as the runner-up score.**

```python
n = int(input("Enter n: "))
scores = list(map(int, input("Enter scores: ").split()))

scores = list(set(scores))   # remove duplicates
scores.sort()

if len(scores) > 1:
    print("Runner-up:", scores[-2])
else:
    print("Runner-up not possible (all scores are same)")
```

### **4) Create a dictionary of n persons where key is name and value is city.**

**a. Display all names**

**b. Display all city names**

**c. Display student names and cities of all students.**

**d. Count the number of students in each city.**


```python
n=int(input("Enter n: "))
d={input("Name: "):input("City: ") for i in range(n)}
print("Names:",list(d.keys()))
print("Cities:",list(d.values()))
print("City Count:",{c:list(d.values()).count(c) for c in set(d.values())})
```

**c. print movies that made a profit.**
